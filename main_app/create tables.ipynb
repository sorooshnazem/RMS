```python
import sqlite3

# Connect to a database (creates the file if it doesn't exist)
conn = sqlite3.connect('restaurant.db')

# Create a cursor object to execute SQL commands
cursor = conn.cursor()

# Create a simple table
cursor.execute('''
    CREATE TABLE IF NOT EXISTS `food` (
	`id_food_id` integer primary key NOT NULL UNIQUE,
	`gn_food_name` TEXT NOT NULL,
	`id_food_category_id` INTEGER NOT NULL,
	`nm_price_number` INTEGER NOT NULL,
	`ds_food_descr` TEXT NOT NULL,
	`id_restaurant_id` INTEGER NOT NULL,
FOREIGN KEY(`id_food_category_id`) REFERENCES `food_category`(`id_food_category_id`),
FOREIGN KEY(`id_restaurant_id`) REFERENCES `restaurant`(`id_restaurant_id`)
)
''')

# Create a simple table
cursor.execute('''
    CREATE TABLE IF NOT EXISTS `restaurant` (
	`id_restaurant_id` integer primary key NOT NULL UNIQUE,
	`gn_restaurant_name` TEXT NOT NULL,
	`ds_restaurant_address_descr` TEXT NOT NULL
)
''')

cursor.execute('''
    CREATE TABLE IF NOT EXISTS `order` (
	`id_order_id` integer primary key NOT NULL UNIQUE,
	`ds_order_descr` TEXT,
	`dt_order_date` REAL NOT NULL,
	`dt_total_price_date` REAL NOT NULL,
	`id_user_id` INTEGER NOT NULL,
    FOREIGN KEY(`id_user_id`) REFERENCES `User`(`id_user_id`)
)
''')

cursor.execute('''
    CREATE TABLE IF NOT EXISTS `order_detail` (
	`id_irder_id` INTEGER NOT NULL,
	`id_food_id` INTEGER NOT NULL,
	`nm_food_ordered_number` INTEGER NOT NULL,
FOREIGN KEY(`id_irder_id`) REFERENCES `order`(`id_order_id`),
FOREIGN KEY(`id_food_id`) REFERENCES `food`(`id_food_id`)
)
''')

cursor.execute('''
    CREATE TABLE IF NOT EXISTS `food_category` (
	`id_food_category_id` integer primary key NOT NULL UNIQUE,
	`gn_food_category_name` TEXT NOT NULL,
	`ds_food_category_descr` TEXT
)
''')

cursor.execute('''
    CREATE TABLE IF NOT EXISTS `User` (
	`id_user_id` integer primary key NOT NULL,
	`gn_surname` TEXT NOT NULL,
	`nm_phone_number` INTEGER NOT NULL
)
''')

# Commit the changes
conn.commit()

# Close the connection
conn.close()
```
